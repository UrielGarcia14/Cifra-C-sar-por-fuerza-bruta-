```python
#Importar las librerias para cargar y descargar archivos.
import os
from google.colab import files
```

```python
#Cargar documentos
doc = files.upload()
#Lectura del archivo
f = open ('eng2.crp','r')
criptograma = f.read()
print("-----Mensaje encriptado-----")
print(criptograma)
```

```python
# Recupera la clave de una cifra César por fuerza bruta 
ASCII = """ !"#$%&\'()*+,-./0123456789:;<=>?@ABCDEFGHIJKLMNOPQRSTUVWXYZ[\\]^_`abcdefghijklmnopqrstuvwxyz{|}"""
# Recorre una a una todas las claves
for clave in range(1,len(ASCII)):
  # Almacena la cadena descifrada
  salida = ''
  for simbolo in criptograma:
    if simbolo in ASCII: 
      pos = ASCII.index(simbolo)
      # Descifra el carácter
      pos = (pos - clave) % len(ASCII)
      # Añade el símbolo descifrado a la cadena
      salida += ASCII[pos]
    # Si hay un espacio u otro carácter no 
    # alfabético lo añade a la cadena sin tocar
    else:
      salida += simbolo
# Imprime en pantalla el resultado completo
  print('------------Clave %d ----------' % clave)
  print(' %s ' % salida)
```

```python
key = int ( input('Clave: '))
# Almacena la cadena descifrada
exit = ''
for simbolo in criptograma:
  if simbolo in ASCII: 
    position = ASCII.index(simbolo)
    # Descifra el carácter
    position = (position - key) % len(ASCII)
    # Añade el símbolo descifrado a la cadena
    exit += ASCII[position]
    # Si hay un espacio u otro carácter no 
    # alfabético lo añade a la cadena sin tocar
  else:
    exit += simbolo
```

```python
#Creando un Archivo txt del archivo decodificado.
Archivo = open ( "eng2.txt" , "w" )
Archivo.write(exit + os.linesep)
Archivo.close ( )
#Descarga del Archivo Decifrado
files.download ('eng2.txt')
```
